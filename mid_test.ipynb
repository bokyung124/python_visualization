##3-4 주어진 문자열의 처음 2개 문자와 마지막 2개 문자로 구성된 문자열을 가져오는 Python 프로그램을 작성하시오.(단, 문자열의 길이가 2보다 작으면 빈 문자열을 반환함)

```python
#아래에 답안을 작성하시오.

def str(string):
  if(len(string)<=2):
    print('')
  else:
    print(string[:2] + string[-2:])

str('hello world')
str('j')
```

##3-5 딕셔너리에서 키가 동일한 값을 합산하여 출력하는 Python 프로그램을 작성하시오. 
*   d1 = {'sook': 100, 'myung': 200, 'univ':300}
*   d2 = {'sook': 300, 'myung': 200, 'women':400}
*   Sample output: Counter({'sook': 400, 'myung': 400, 'women': 400, 'univ': 300})

```python
#아래에 답안을 작성하시오.
from collections import Counter

d1 = {'sook': 100, 'myung': 200, 'univ':300}
d2 = {'sook': 300, 'myung': 200, 'women':400}

Counter = dict(Counter(d1)+Counter(d2))
Counter
```

##3-8 문자열을 받아서 대문자와 소문자의 수를 각각 세는 Python 함수를 작성하시오.
*   Sample List : 'Sookmyung Women's University'
*   Expected Output :
*   No. of Upper case characters : 3
*   No. of Lower case Characters : 22

```python
string = 'Sookmyung Women\'s University'

def count(str):
  upper = sum(c.isupper() for c in str)
  lower = sum(c.islower() for c in str)
  print('No. of Upper case characters :', upper)
  print('No. of Lower case Characters :', lower)

count(string)
```
